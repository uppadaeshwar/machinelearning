```python
import pandas as pd
```

```python
mobile_price = pd.read_csv("/content/mobile phone price prediction.csv")
```

```python
mobile_price.head()
```

```python
mobile_price.describe()
```

```python
len(mobile_price)
```

```python
mobile_price.isna().sum()
```

```python
mobile_price.dropna(inplace=True)
```

```python
len(mobile_price)
```

```python
mobile_price.drop(['Unnamed: 0','Name'], axis=1, inplace=True)
mobile_price.info()
```

```python
mobile_price.head(3)
```

```python
int_columns = ["Ram", "Battery", "Display", "External_Memory", "Android_version", "Price", "Inbuilt_memory", "fast_charging", "Screen_resolution"]
mobile_price['Price'] = mobile_price['Price'].str.replace(',', '').astype(int)
```

```python
for column in int_columns:
    # Ensure the column is treated as a string before extracting numeric parts
    mobile_price[column] = mobile_price[column].astype(str).str.extract('(\d+)').astype(float)
    if mobile_price[column].isnull().any():
        mode_value = mobile_price[column].mode()[0]
        mobile_price[column] = mobile_price[column].fillna(mode_value)

mobile_price.info()
```

```python
mobile_price.info()
```

```python
mobile_price.head()
```

```python
float_col = ["Rating", "Display"]
int_col = ["Spec_score", "Ram", "Battery", "Android_version", "Price", "Inbuilt_memory", "fast_charging", "Screen_resolution"]
mobile_price[float_col] = mobile_price[float_col].astype(float)
mobile_price[int_col] = mobile_price[int_col].astype(int)
```

```python
mobile_price.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
obj_cols = mobile_price.select_dtypes(include='object').columns
obj_cols
```

```python
for col in obj_cols:
  mobile_price[col] = le.fit_transform(mobile_price[col])
```

```python
mobile_price.head(3)
```

```python
mobile_price.corr()
```

```python
from sklearn.model_selection import train_test_split
X = mobile_price.drop("Price", axis=1)
y = mobile_price["Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
y_preds = model.predict(X_test)
```

```python
print(mean_absolute_error(y_test, y_preds))
```
